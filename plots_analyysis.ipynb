```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load iris dataset
iris = sns.load_dataset('iris')

# Correlation matrix only on numeric columns
corr = iris.select_dtypes(include=['float64', 'int64']).corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap - Iris Dataset")
plt.show()
```

```python
df = pd.DataFrame({'day': pd.date_range(start="2023-01-01", periods=10, freq='D'),
                   'sales':[100,110,105,115,120,125,130,140,135,150]})
df['rolling'] = df['sales'].rolling(3).mean()

plt.plot(df['day'], df['sales'], label="Sales")
plt.plot(df['day'], df['rolling'], label="3-Day Avg")
plt.legend()
plt.show()
```
